```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Example: Generating random data (replace with your actual data)
np.random.seed(42)
X = np.random.rand(100, 2)  # 100 samples with 2 features
X[:10]

# If you have a dataset in a CSV file
# df = pd.read_csv('your_dataset.csv')
# X = df[['feature1', 'feature2']].values  # select your relevant features
```

```python
# Initialize KMeans with a specified number of clusters (K)
kmeans = KMeans(n_clusters=3, random_state=42)  # K=3 here, adjust as needed

# Fit the model to the data
kmeans.fit(X)
```

```python
# Get the cluster labels (which cluster each point belongs to)
labels = kmeans.labels_

# Get the coordinates of the centroids
centroids = kmeans.cluster_centers_
```

```python
# Show the unique cluster labels
print('Unique cluster labels:', np.unique(labels))
print('Centroids:', centroids)
```

```python
# Scatter plot of data points
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', label='Data Points')

# Plot the centroids
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', marker='X', label='Centroids')

plt.title('K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
inertia = []
K = range(1, 10)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)  # Sum of squared distances

# Plot the Elbow Curve
plt.plot(K, inertia, 'bx-')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Inertia (Sum of squared distances)')
plt.title('Elbow Method For Optimal K')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Example dataset with 5 features
X = np.array([
    [25, 50000, 16, 1, 3],   # [age, income, education, political view, traffic deaths]
    [40, 60000, 18, 2, 4],
    [35, 70000, 20, 1, 5],
    [45, 80000, 15, 3, 2],
    [30, 55000, 14, 2, 3],
    [50, 90000, 19, 1, 5],
    [56, 32695, 14, 1, 2],
[46, 78190, 10, 4, 2],
[32, 35258, 10, 1, 1],
[60, 117538, 12, 2, 1],
[25, 69504, 11, 1, 1],
[38, 63159, 14, 4, 3],
[56, 43986, 21, 4, 5],
[36, 91858, 19, 4, 2],
[40, 42666, 15, 4, 2],
[28, 68660, 16, 4, 3],
[48, 119112, 19, 1, 2],
[27, 41130, 17, 2, 5],
[36, 117508, 15, 4, 5],
[49, 90566, 21, 1, 1],
[18, 96703, 17, 2, 5],
[22, 57192, 21, 2, 1],
[62, 31062, 14, 3, 4],
[21, 114076, 17, 1, 2],
[33, 119780, 19, 1, 2],
[41, 51545, 21, 4, 1]
])

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

# Reduce dimensionality to 2D using PCA for visualization
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

# Visualize the clusters
plt.figure(figsize=(8, 6))

# Plot each cluster with a different color
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k')

# Plot the centroids
centroids = pca.transform(kmeans.cluster_centers_)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=100, label='Centroids')

plt.title("K-means Clusters Visualized in 2D using PCA")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.show()
```
